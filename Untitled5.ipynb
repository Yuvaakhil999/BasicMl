**HEAT MAP**

```python
def null_age(cols):
  Age = cols[0]
  Pclass = cols[1]

  if pd.isnull(Age):

    if Pclass == 1:
      return 47

    elif Pclass == 2:
      return 39

    else:
        return 44

  else:
      return Age
```

```python
travel['Age'] = travel[['Age','Pclass']].apply(null_age,axis=1)
```

```python
sns.heatmap(travel.isnull())
```

**CLEANING DATASET**

```python
travel.drop('Cabin',axis=1,inplace=True)
```
